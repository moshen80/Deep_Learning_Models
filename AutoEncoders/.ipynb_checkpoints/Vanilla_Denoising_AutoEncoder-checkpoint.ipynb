```python
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.datasets import mnist
from tensorflow.keras.layers import Conv2D, MaxPooling2D, UpSampling2D
from tensorflow.keras.models import Sequential
```

### Loading dataset
Loading, Scaling and Reshaping MNIST image dataset

```python
(x_train, _), (x_test, _) = mnist.load_data()

x_train = x_train.astype('float32') / 255.
x_test = x_test.astype('float32') / 255.
x_train = np.reshape(x_train, (len(x_train), 28, 28, 1))
x_test = np.reshape(x_test, (len(x_test), 28, 28, 1))
```

### Adding Noise to the traning dataset
First, we add a random nomarl noise to the input images and the we clip the noisy images to be between 0,1

```python
noise_std = 0.5
x_train_noisy = x_train + noise_std * np.random.normal(loc=0.0, scale=1.0, size=x_train.shape) 
x_test_noisy = x_test + noise_std * np.random.normal(loc=0.0, scale=1.0, size=x_test.shape) 
```

```python
x_train_noisy = np.clip(x_train_noisy, 0., 1.)
x_test_noisy = np.clip(x_test_noisy, 0., 1.)
```

### Visualizing some of original and noisy images

```python
fig, ax = plt.subplots(2,10,figsize=(20, 4))
for i in range(0,10):
    ax[0,i].imshow(x_test[i].reshape(28, 28), cmap="binary")
    ax[1,i].imshow(x_test_noisy[i].reshape(28, 28), cmap="binary")
plt.show()
```

### Building the the AutoEncoder Model using Keras

```python
# Encoder
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=(28, 28, 1)))
model.add(MaxPooling2D((2, 2), padding='same'))
model.add(Conv2D(8, (3, 3), activation='relu', padding='same'))
model.add(MaxPooling2D((2, 2), padding='same'))
model.add(Conv2D(8, (3, 3), activation='relu', padding='same'))
 

model.add(MaxPooling2D((2, 2), padding='same'))

## Decoder 
model.add(Conv2D(8, (3, 3), activation='relu', padding='same'))
model.add(UpSampling2D((2, 2)))
model.add(Conv2D(8, (3, 3), activation='relu', padding='same'))
model.add(UpSampling2D((2, 2)))
model.add(Conv2D(32, (3, 3), activation='relu'))
model.add(UpSampling2D((2, 2)))
model.add(Conv2D(1, (3, 3), activation='relu', padding='same'))

model.compile(optimizer='adam', loss='mean_squared_error')

model.summary()
```

```python
# Fitting the Model
model.fit(x_train_noisy, x_train, epochs=10, batch_size=256, shuffle=True, 
          validation_data=(x_test_noisy, x_test))
```

```python
#Evaluating the Model (MSE)
model.evaluate(x_test_noisy, x_test)
```

```python
# Saving the Model
# model.save('denoising_autoencoder.model')
```

```python
no_noise_img = model.predict(x_test_noisy)
```

### Plotting Original, Noisy and Reconstructed Images

```python
fig, ax = plt.subplots(3,10,figsize=(20, 6))
for i in range(0,10):
    ax[0,i].imshow(x_test[i].reshape(28, 28), cmap="binary")
    ax[1,i].imshow(x_test_noisy[i].reshape(28, 28), cmap="binary")
    ax[2,i].imshow(no_noise_img[i].reshape(28, 28), cmap="binary")
plt.show()
```
